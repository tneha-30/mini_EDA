```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt      #visualisation
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\USER\Desktop\EDA_project\retail_data.csv")
df.info()
```

```python
df.head() # prints first 5 rows
# df.head(7) will print 7 rows
```

```python
df[['date', 'month', 'year']] = df['order_date'].str.split('/', expand=True)

print("\nDataFrame after splitting 'order_date' into 'date','month' and 'year':")
df.info()
```

```python
# Removing unnecessary values
df.drop(['ship_date','country', 'order_date'], axis=1, inplace=True)
```

```python
df.info()
```

```python
# To Check NULL values
pd.isnull(df).sum()

# If null values are found then
# Example of handling missing values
# Fill missing values with mean (for numerical columns)
# df.fillna(df.mean(), inplace=True)

# Drop rows with missing values (if necessary)
# df.dropna(inplace=True)
```

```python
# Removing duplicates

duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)
```

```python
# Drop duplicate rows
df = df.drop_duplicates()
```

```python
duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)
```

## Exploratory Data Analysis

```python
a = sns.countplot(x='region', data=df)

for i in a.containers:
    a.bar_label(i)
```

```python
a = sns.countplot(x='region', hue='category', data=df)

for i in a.containers:
    a.bar_label(i)
```

```python
a = sns.countplot(x='year', hue='category', data=df)

for i in a.containers:
    a.bar_label(i)
```

```python
sales_state = df.groupby(['state'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False).head(15)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x='state', y='quantity')
```

```python
sales_subcategory = df.groupby(['subcategory'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data= sales_subcategory, x='subcategory', y='quantity')
```

```python
sales_year = df.groupby(['year'],as_index=False)['sales'].sum().sort_values(by='year')
sns.set(rc={'figure.figsize':(3, 2)})
sns.barplot(data= sales_year, x='year', y='sales')
```

```python
df['profit_percentage'] = (df['profit'] / df['sales']) * 100
# df['profit_percentage'] = (df['profit_margin']) * 100
```

```python
print(df)
```

```python
sns.lineplot(data=df, x='year', y='profit_percentage', marker='o')
plt.title('Profit Percentage by Year')
plt.ylabel('Profit Percentage (%)')
plt.xlabel('Year')
plt.show()
```

```python
top_products = df.groupby(['product_name'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False).head(5)
sns.set(rc={'figure.figsize':(12, 5)})
sns.barplot(data= top_products, x='product_name', y='quantity')
```

```python
city_top_products = df.groupby(['city'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False).head(5)
sns.set(rc={'figure.figsize':(12, 5)})
sns.barplot(data= city_top_products, x='city', y='quantity')
```

```python
grouped_df = df.groupby(['city', 'product_name'], as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False).head(15)

sns.set(rc={'figure.figsize': (20, 5)})  # Set the figure size
sns.barplot(data=grouped_df, x='city', y='quantity', hue='product_name', palette='Paired')

# Add titles and labels
plt.title('Top 5 City-Product Combinations by Quantity Sold')
plt.ylabel('Total Quantity Sold')
plt.xlabel('City')

# Show the plot
plt.show()
```
